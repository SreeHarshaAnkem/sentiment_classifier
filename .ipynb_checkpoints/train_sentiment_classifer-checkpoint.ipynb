```python
class Attention(L.Layer):

    def __init__(self):
        super(Attention, self).__init__()


    def build(self, input_shape):
        # Create a trainable weight variable for this layer.
        self.mlp = L.Dense(units=2*LSTM_UNITS,
                           trainable=True, activation="tanh")
        self.context_vector = L.Dense(
            units=1, trainable=True, activation="linear")
        super(Attention, self).build(input_shape)  # Be sure to call this at the end

    def call(self, x):
        mlp_out = self.mlp(x)
        context = self.context_vector(mlp_out)
        scores = L.Flatten()(context)
        weights = L.Activation(activation="softmax")(scores)
        weights = L.RepeatVector(n=2*LSTM_UNITS)(weights)
        weights = L.Permute((2, 1))(weights)
        return weights
```

```python
class Attention:
    '''
    Reference: https://keras.io/layers/writing-your-own-keras-layers/
    '''

    def __init__(self):
        self.mlp = L.Dense(units=2*LSTM_UNITS,
                           trainable=True, activation="tanh")
        self.context_vector = L.Dense(
            units=1, trainable=True, activation="linear")

    def __call__(self, x):
        mlp_out = self.mlp(x)
        context = self.context_vector(mlp_out)
        scores = L.Flatten()(context)
        weights = L.Activation(activation="softmax")(scores)
        weights = L.RepeatVector(n=2*LSTM_UNITS)(weights)
        weights = L.Permute((2, 1))(weights)
        return weights
```

```python
def model_architecture(mode = "train"):
    # Input
    review_input = L.Input(shape=(MAX_TOKENS_PER_REVIEW,), name="review_input")
    # Embeddings
    embeddings = L.Embedding(input_dim=VOCAB_SIZE,
                             output_dim=EMBEDDING_SIZE)(review_input)
    # LSTM
    all_hidden_states = L.Bidirectional(
        L.LSTM(units=LSTM_UNITS, return_sequences=True))(embeddings)

    # Attention
    attention_weights = Attention()(all_hidden_states)
    all_hidden_states = L.Multiply()([all_hidden_states, attention_weights])

    # Weighted Hidden states of all timesteps
    effective_state = L.Lambda(lambda states: K.sum(
        states, axis=1))(all_hidden_states)

    # Fully Connected Layer
   # mlp_out = L.Dense(units=DENSE_UNITS, activation="relu")(effective_state)

    # Readout Layer
    prediction = L.Dense(units=1, activation="sigmoid")(effective_state)

    # Define Sequential Model
    model = Model(inputs=[review_input], outputs=[prediction])
    if mode == "interpretation":
        model_attn = Model(inputs = [review_input], outputs = [attention_weights])
    else:
        return model
```

```python
if __name__ == "__main__":

    EMBEDDING_SIZE = 4
    VOCAB_SIZE = 5537
    LSTM_UNITS = 4
    DENSE_UNITS = 8

    df = pd.read_pickle("data/processed_data/train_processed_data.pkl")
    X = df.loc[:, ["corpus", "processed_idx"]]
    y = df.loc[:, "sentiment"]

    train_X, test_X, train_y, test_y = train_test_split(
        X, y, stratify=y, test_size=0.2, random_state=42)
    print(train_X.shape, train_y.shape, test_X.shape, test_y.shape)

    MAX_TOKENS_PER_REVIEW = int(train_X["processed_idx"].apply(len).quantile(q=0.75))
    train_X_padded = pad_sequences(
        train_X["processed_idx"].values.tolist(), maxlen=MAX_TOKENS_PER_REVIEW)
    test_X_padded = pad_sequences(
        test_X["processed_idx"].values.tolist(), maxlen=MAX_TOKENS_PER_REVIEW)

 # Input
    review_input = L.Input(shape=(MAX_TOKENS_PER_REVIEW,), name="review_input")
    # Embeddings
    embeddings = L.Embedding(input_dim=VOCAB_SIZE,
                             output_dim=EMBEDDING_SIZE)(review_input)
    embeddings = L.SpatialDropout1D(rate=0.3)(embeddings)
    # LSTM
    all_hidden_states = L.Bidirectional(
        L.LSTM(units=LSTM_UNITS, return_sequences=True))(embeddings)

    # Attention
    attention_weights = Attention()(all_hidden_states)
    all_hidden_states = L.Multiply()([all_hidden_states, attention_weights])

    # Weighted Hidden states of all timesteps
    effective_state = L.Lambda(lambda states: K.sum(
        states, axis=1))(all_hidden_states)

    # Fully Connected Layer
   # mlp_out = L.Dense(units=DENSE_UNITS, activation="relu")(effective_state)

    # Readout Layer
    prediction = L.Dense(units=1, activation="sigmoid")(effective_state)

    # Define Sequential Model
    model = Model(inputs=[review_input], outputs=[prediction])

    print(model.summary())
    es = EarlyStopping(monitor='val_loss', min_delta=0, patience=0, verbose=0,
                       mode='auto', baseline=None, restore_best_weights=False)
    mc = ModelCheckpoint(filepath="data/processed_data/weights.h5", monitor='val_loss', verbose=0,
                         save_best_only=False, save_weights_only=True, mode='auto', period=1)
    model.compile(loss="binary_crossentropy", optimizer=Adam(
        lr=7e-4), metrics=["accuracy"])
    hist = model.fit(train_X_padded, train_y, epochs=50,
                     validation_split=0.2, callbacks=[es, mc], batch_size = 128)
```
